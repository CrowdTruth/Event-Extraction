```python
import pandas as pd
import numpy as np
```

```python
# number of documents in the dataset
gold = pd.read_csv("../../data/TempEval3-data/TE3-Gold_sentences.csv", skip_blank_lines=True)
goldDocs = gold['Doc Id'].value_counts()
print("#docs in TempEval-3 Gold dataset: ", len(goldDocs))

# number of documents in the dataset
platinum = pd.read_csv("../../data/TempEval3-data/TE3-Platinum_sentences.csv", skip_blank_lines=True)
platinumDocs = platinum['Doc Id'].value_counts()
print("#docs in TempEval-3 Platinum dataset: ", len(platinumDocs))
```

```python
# number of sentences in the dataset
gold = pd.read_csv("../../data/TempEval3-data/TE3-Gold_sentences.csv", skip_blank_lines=True)
print("#sent in TempEval-3 Gold dataset: ", len(gold))

# number of sentences in the dataset
platinum = pd.read_csv("../../data/TempEval3-data/TE3-Platinum_sentences.csv", skip_blank_lines=True)
print("#sent in TempEval-3 Platinum dataset: ", len(platinum))
```

```python
# number of tokens in the dataset
gold = pd.read_csv("../../data/TempEval3-data/TE3-Gold_tokens.csv", skip_blank_lines=True)
print("#tokens in TempEval-3 Gold dataset: ", len(gold))

# number of tokens in the dataset
platinum = pd.read_csv("../../data/TempEval3-data/TE3-Platinum_tokens.csv", skip_blank_lines=True)
print("#tokens in TempEval-3 Platinum dataset: ", len(platinum))
```

```python
# number of events in the dataset
goldevents = pd.read_csv("../../data/TempEval3-data/TE3-Gold_events.csv", skip_blank_lines=True)
print("#events in TempEval-3 Gold dataset: ", len(goldevents))

# number of events in the dataset
platinumevents = pd.read_csv("../../data/TempEval3-data/TE3-Platinum_events.csv", skip_blank_lines=True)
print("#events in TempEval-3 Platinum dataset: ", len(platinumevents))
```

```python
# number of time expressions in the dataset
goldtime = pd.read_csv("../../data/TempEval3-data/TE3-Gold_time.csv", skip_blank_lines=True)
print("#time expressions in TempEval-3 Gold dataset: ", len(goldtime))

# number of time expressions in the dataset
platinumtime = pd.read_csv("../../data/TempEval3-data/TE3-Platinum_time.csv", skip_blank_lines=True)
print("#time expressions in TempEval-3 Platinum: ", len(platinumtime))
```

```python
annotatedSentencesGold = goldevents.groupby(["Doc Id", "Sentence Id"]).size().reset_index(name="Count")
print("# annotated sentences with events in the TempEval-3 Gold dataset: ", len(annotatedSentencesGold))
print("average number of events per sentences in the TempEval-3 Gold dataset: ", float(len(goldevents)) / float(len(annotatedSentencesGold)))

annotatedSentencesPlatinum = platinumevents.groupby(["Doc Id", "Sentence Id"]).size().reset_index(name="Count")
print("# annotated sentences with events in the TempEval-3 Platinum dataset: ", len(annotatedSentencesPlatinum))
print("average number of events per sentences in the TempEval-3 Platinum dataset: ", float(len(platinumevents)) / float(len(annotatedSentencesPlatinum)))
```

```python
# number of annotated sentences with time expressions and average number of time expressions per sentence
annotatedSentencesGold = goldtime.groupby(["Doc Id", "Sentence Id"]).size().reset_index(name="Count")
print("# annotated sentences with time expressions in the TempEval-3 Gold dataset: ", len(annotatedSentencesGold))
print("average number of time expressions per sentences in the TempEval-3 Gold dataset: ", float(len(goldtime)) / float(len(annotatedSentencesGold)))

annotatedSentencesPlatinum = platinumtime.groupby(["Doc Id", "Sentence Id"]).size().reset_index(name="Count")
print("# annotated sentences with time expressions in the TempEval-3 Platinum dataset: ",len(annotatedSentencesPlatinum))
print("average number of time expressions per sentences in the TempEval-3 Platinum dataset: ", float(len(platinumtime)) / float(len(annotatedSentencesPlatinum)))
```

```python
# number of sentences not annotated with events 
goldsent = pd.read_csv("../../data/TempEval3-data/TE3-Gold_sentences.csv", skip_blank_lines=True)
platinumsent = pd.read_csv("../../data/TempEval3-data/TE3-Platinum_sentences.csv", skip_blank_lines=True)

annotatedSentencesGold = goldevents.groupby(["Doc Id", "Sentence Id"]).size().reset_index(name="Count")
print("# sentences not annotated with events in the TempEval-3 Gold dataset: ", len(goldsent) - len(annotatedSentencesGold))

annotatedSentencesPlatinum = platinumevents.groupby(["Doc Id", "Sentence Id"]).size().reset_index(name="Count")
print("# sentences not annotated with events in the TempEval-3 Gold dataset: ", len(platinumsent) - len(annotatedSentencesPlatinum))
```

```python
# number of sentences not annotated with time expressions 
goldsent = pd.read_csv("../../data/TempEval3-data/TE3-Gold_sentences.csv", skip_blank_lines=True)
platinumsent = pd.read_csv("../../data/TempEval3-data/TE3-Platinum_sentences.csv", skip_blank_lines=True)

annotatedSentencesGold = goldtime.groupby(["Doc Id", "Sentence Id"]).size().reset_index(name="Count")
print("# sentences not annotated with time expressions in the TempEval-3 Gold dataset: ", len(goldsent) - len(annotatedSentencesGold))

annotatedSentencesPlatinum = platinumtime.groupby(["Doc Id", "Sentence Id"]).size().reset_index(name="Count")
print("# sentences not annotated with time expressions in the TempEval-3 Gold dataset: ", len(platinumsent) - len(annotatedSentencesPlatinum))
```

```python
# distribution of POS Tags for the annotated events
goldPOSTagsHisto = pd.DataFrame(goldevents['Pos'].value_counts().reset_index())
goldPOSTagsHisto.columns = ['POS Tag', 'Count']
print("POS tags distribution for the annotated events in TempEval-3 Gold dataset")
print(goldPOSTagsHisto)
print("\n")

platinumPOSTagsHisto = pd.DataFrame(platinumevents['Pos'].value_counts().reset_index())
platinumPOSTagsHisto.columns = ['POS Tag', 'Count']
print("POS tags distribution for the annotated events in TempEval-3 Platinum dataset")
print(platinumPOSTagsHisto)
```

```python
# distribution of POS Tags for the annotated time expressions
goldPOSTagsHisto = pd.DataFrame(goldtime['Pos'].value_counts().reset_index())
goldPOSTagsHisto.columns = ['POS Tag', 'Count']
print("POS tags distribution for the annotated time expressions in TempEval-3 Gold dataset")
print(goldPOSTagsHisto)
print("\n")

platinumPOSTagsHisto = pd.DataFrame(platinumtime['Pos'].value_counts().reset_index())
platinumPOSTagsHisto.columns = ['POS Tag', 'Count']
print("POS tags distribution for the annotated time expressions in TempEval-3 Platinum dataset")
print(platinumPOSTagsHisto)
```

```python
# average number of times an event (as a combination of token and POS tag) is annotated in the dataset 
goldUniqueToken = goldevents.groupby(["Lowercase Token", "Pos"]).size().reset_index(name="Count")
goldUniqueToken.columns = ['Lowercase Token', 'POS Tag', 'Count']
print("avg number of times an event token is annotated in the TempEval-3 Gold dataset: ", goldUniqueToken["Count"].mean())
print("unique number of event tokens annotated in the TempEval-3 Gold dataset: ", len(goldUniqueToken))

goldUniqueLemma = goldevents.groupby(["Lowercase Lemma Stanford", "Pos"]).size().reset_index(name="Count")
goldUniqueLemma.columns = ['Lowercase Lemma Stanford', 'POS Tag', 'Count']
print("avg number of times an event lemma is annotated in the TempEval-3 Gold dataset: ", goldUniqueLemma["Count"].mean())
print("unique number of event lemmas annotated in the TempEval-3 Gold dataset: ", len(goldUniqueLemma))

platinumUniqueToken = platinumevents.groupby(["Lowercase Token", "Pos"]).size().reset_index(name="Count")
platinumUniqueToken.columns = ['Lowercase Token', 'POS Tag', 'Count']
print("avg number of times an event token is annotated in the TempEval-3 Platinum dataset: ", platinumUniqueToken["Count"].mean())
print("unique number of event tokens annotated in the TempEval-3 Platinum dataset: ", len(platinumUniqueToken))

platinumUniqueLemma = platinumevents.groupby(["Lowercase Lemma Stanford", "Pos"]).size().reset_index(name="Count")
platinumUniqueLemma.columns = ['Lowercase Lemma Stanford', 'POS Tag', 'Count']
print("avg number of times an event lemma is annotated in the TempEval-3 Platinum dataset: ", platinumUniqueLemma["Count"].mean())
print("unique number of event lemmas annotated in the TempEval-3 Platinum dataset: ", len(platinumUniqueLemma))
```
