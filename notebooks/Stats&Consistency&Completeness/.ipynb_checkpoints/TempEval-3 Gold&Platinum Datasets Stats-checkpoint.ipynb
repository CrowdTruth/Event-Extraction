```python
# number of events in the dataset
goldEvents = pd.read_csv("../TempEval3-data/TBAQ-cleaned_events.csv", skip_blank_lines=True)
platinumEvents = pd.read_csv("../TempEval3-data/TE3-Platinum_events.csv", skip_blank_lines=True)

goldTokens = pd.read_csv("../TempEval3-data/TBAQ-cleaned_tokens.csv", skip_blank_lines=True)
platinumTokens = pd.read_csv("../TempEval3-data/TE3-Platinum_tokens.csv", skip_blank_lines=True)

goldUniqueToken = goldTokens.groupby(["Lowercase Token", "POS Tag"]).size().reset_index(name="Count")
goldUniqueToken.columns = ['Lowercase Token', 'POS Tag', 'Count']
goldUniqueLemma = goldTokens.groupby(["Lowercase Lemma Stanford", "POS Tag"]).size().reset_index(name="Count")
goldUniqueLemma.columns = ['Lowercase Lemma Stanford', 'POS Tag', 'Count']

goldUniqueEvent = goldEvents.groupby(["LowercaseToken", "Pos"]).size().reset_index(name="Count")
goldUniqueEvent.columns = ['Lowercase Token', 'POS Tag', 'Count']
goldUniqueLemmaEvent = goldEvents.groupby(["LowercaseLemmaStanford", "Pos"]).size().reset_index(name="Count")
goldUniqueLemmaEvent.columns = ['Lowercase Lemma Stanford', 'POS Tag', 'Count']

goldEventStats = pd.merge(goldUniqueEvent, goldUniqueToken, on=['Lowercase Token', 'POS Tag'], how='left')
goldEventLemmaStats = pd.merge(goldUniqueLemmaEvent, goldUniqueLemma, on=['Lowercase Lemma Stanford', 'POS Tag'], how='left')
goldEventStats.columns = ['Lowercase Token', 'POS Tag', 'Count Event', 'Count Token']
goldEventLemmaStats.columns = ['Lowercase Lemma Stanford', 'POS Tag', 'Count Event Lemma', 'Count Token Lemma']


platinumUniqueToken = platinumTokens.groupby(["Lowercase Token", "POS Tag"]).size().reset_index(name="Count")
platinumUniqueToken.columns = ['Lowercase Token', 'POS Tag', 'Count']
platinumUniqueLemma = platinumTokens.groupby(["Lowercase Lemma Stanford", "POS Tag"]).size().reset_index(name="Count")
platinumUniqueLemma.columns = ['Lowercase Lemma Stanford', 'POS Tag', 'Count']

platinumUniqueEvent = platinumEvents.groupby(["LowercaseToken", "Pos"]).size().reset_index(name="Count")
platinumUniqueEvent.columns = ['Lowercase Token', 'POS Tag', 'Count']
platinumUniqueLemmaEvent = platinumEvents.groupby(["LowercaseLemmaStanford", "Pos"]).size().reset_index(name="Count")
platinumUniqueLemmaEvent.columns = ['Lowercase Lemma Stanford', 'POS Tag', 'Count']


platinumEventStats = pd.merge(platinumUniqueEvent, platinumUniqueToken, on=['Lowercase Token', 'POS Tag'], how='left')
platinumEventLemmaStats = pd.merge(platinumUniqueLemmaEvent, platinumUniqueLemma, on=['Lowercase Lemma Stanford', 'POS Tag'], how='left')
platinumEventStats.columns = ['Lowercase Token', 'POS Tag', 'Count Event', 'Count Token']
platinumEventLemmaStats.columns = ['Lowercase Lemma Stanford', 'POS Tag', 'Count Event Lemma', 'Count Token Lemma']

```

```python
#print "Total number of events: ", sum(goldEventStats["Count Event"])
#print "Total number of event occurrences in texts: ", sum(goldEventStats["Count Token"])
totalInconsistencies = sum(goldEventStats["Count Token"]) - sum(goldEventStats["Count Event"])
totalInconsistenciesPer = totalInconsistencies * 100.0 / sum(goldEventStats["Count Token"])
print "Total number of inconsistencies ALL: ", totalInconsistencies, totalInconsistenciesPer 

distinctInconsistencies = len(goldEventStats[goldEventStats["Count Event"] != goldEventStats["Count Token"]])
distinctInconsistenciesPer = distinctInconsistencies * 100.0 / len(goldEventStats.index)
print "Distinct number of inconsistencies ALL:", distinctInconsistencies, distinctInconsistenciesPer
print "--------------------------"


eventVerbs = goldEventStats[goldEventStats["POS Tag"] == 'VB']
totalInconsistencies = sum(eventVerbs["Count Token"]) - sum(eventVerbs["Count Event"])
totalInconsistenciesPer = totalInconsistencies * 100.0 / sum(eventVerbs["Count Token"])
print "Total number of inconsistencies VB: ", totalInconsistencies, totalInconsistenciesPer 

distinctInconsistencies = len(eventVerbs[eventVerbs["Count Event"] != eventVerbs["Count Token"]])
distinctInconsistenciesPer = distinctInconsistencies * 100.0 / len(eventVerbs.index)
print "Distinct number of inconsistencies VB:", distinctInconsistencies, distinctInconsistenciesPer
print "--------------------------"


eventNouns = goldEventStats[goldEventStats["POS Tag"] == 'NN']
totalInconsistencies = sum(eventNouns["Count Token"]) - sum(eventNouns["Count Event"])
totalInconsistenciesPer = totalInconsistencies * 100.0 / sum(eventNouns["Count Token"])
print "Total number of inconsistencies NN: ", totalInconsistencies, totalInconsistenciesPer 

distinctInconsistencies = len(eventNouns[eventNouns["Count Event"] != eventNouns["Count Token"]])
distinctInconsistenciesPer = distinctInconsistencies * 100.0 / len(eventNouns.index)
print "Distinct number of inconsistencies NN:", distinctInconsistencies, distinctInconsistenciesPer
print "--------------------------"


eventPrep = goldEventStats[goldEventStats["POS Tag"] == 'IN']
totalInconsistencies = sum(eventPrep["Count Token"]) - sum(eventPrep["Count Event"])
totalInconsistenciesPer = totalInconsistencies * 100.0 / sum(eventPrep["Count Token"])
print "Total number of inconsistencies IN: ", totalInconsistencies, totalInconsistenciesPer 

distinctInconsistencies = len(eventPrep[eventPrep["Count Event"] != eventPrep["Count Token"]])
distinctInconsistenciesPer = distinctInconsistencies * 100.0 / len(eventPrep.index)
print "Distinct number of inconsistencies IN:", distinctInconsistencies, distinctInconsistenciesPer
print "--------------------------"



eventAdj = goldEventStats[goldEventStats["POS Tag"] == 'JJ']
totalInconsistencies = sum(eventAdj["Count Token"]) - sum(eventAdj["Count Event"])
totalInconsistenciesPer = totalInconsistencies * 100.0 / sum(eventAdj["Count Token"])
print "Total number of inconsistencies JJ: ", totalInconsistencies, totalInconsistenciesPer 

distinctInconsistencies = len(eventAdj[eventAdj["Count Event"] != eventAdj["Count Token"]])
distinctInconsistenciesPer = distinctInconsistencies * 100.0 / len(eventAdj.index)
print "Distinct number of inconsistencies JJ:", distinctInconsistencies, distinctInconsistenciesPer
print "--------------------------"



eventAdv = goldEventStats[goldEventStats["POS Tag"] == 'RB']
totalInconsistencies = sum(eventAdv["Count Token"]) - sum(eventAdv["Count Event"])
totalInconsistenciesPer = totalInconsistencies * 100.0 / sum(eventAdv["Count Token"])
print "Total number of inconsistencies RB: ", totalInconsistencies, totalInconsistenciesPer 

distinctInconsistencies = len(eventAdv[eventAdv["Count Event"] != eventAdv["Count Token"]])
distinctInconsistenciesPer = distinctInconsistencies * 100.0 / len(eventAdv.index)
print "Distinct number of inconsistencies RB:", distinctInconsistencies, distinctInconsistenciesPer
print "--------------------------"
```
