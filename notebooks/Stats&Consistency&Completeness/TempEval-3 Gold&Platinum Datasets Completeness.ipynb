```python
# check for every token in a time expression how many times it appears in the dataset vs how many times it is annotated as time expression
from itertools import groupby

goldUniqueToken = goldTokens.groupby(["Token POS"]).size().reset_index(name="Count")
goldUniqueToken.columns = ['Token POS', 'Count']
goldUniqueLemma = goldTokens.groupby(["Lemma POS"]).size().reset_index(name="Count")
goldUniqueLemma.columns = ['Lemma POS', 'Count']

goldtimelist = list(goldTime["Token POS"].str.split(' ', expand=True).stack())
counts = pd.Series(goldtimelist).value_counts()
goldUniqueTime = counts.to_frame().reset_index() 
goldUniqueTime.columns=["Token POS", 'Count']
goldtimelist = list(goldTime["Lemma POS"].str.split(' ', expand=True).stack())
counts = pd.Series(goldtimelist).value_counts()
goldUniqueLemmaTime = counts.to_frame().reset_index() 
goldUniqueLemmaTime.columns=["Lemma POS", 'Count']

goldTimeStats = pd.merge(goldUniqueTime, goldUniqueToken, on=['Token POS'], how='left')
colpos = goldTimeStats["Token POS"].apply(lambda x:x.split("-")[-1])
goldTimeStats = goldTimeStats.join(colpos.to_frame(name="POS"))
goldTimeLemmaStats = pd.merge(goldUniqueLemmaTime, goldUniqueLemma, on=['Lemma POS'], how='left')
colpos = goldTimeLemmaStats["Lemma POS"].apply(lambda x:x.split("-")[-1])
goldTimeLemmaStats = goldTimeLemmaStats.join(colpos.to_frame(name="POS"))
goldTimeStats.columns = ['Token POS', 'Count Time', 'Count Token', "POS"]
goldTimeLemmaStats.columns = ['Lemma POS', 'Count Time Lemma', 'Count Token Lemma', "POS"]
goldTimeStats = goldTimeStats[(goldTimeStats["POS"] == "NN") | (goldTimeStats["POS"] == "RB") | (goldTimeStats["POS"] == "JJ")]
goldTimeLemmaStats = goldTimeLemmaStats[(goldTimeLemmaStats["POS"] == "NN") | (goldTimeLemmaStats["POS"] == "RB") | (goldTimeLemmaStats["POS"] == "JJ")]


platinumUniqueToken = platinumTokens.groupby(["Token POS"]).size().reset_index(name="Count")
platinumUniqueToken.columns = ['Token POS', 'Count']
platinumUniqueLemma = platinumTokens.groupby(["Lemma POS"]).size().reset_index(name="Count")
platinumUniqueLemma.columns = ['Lemma POS', 'Count']

platinumtimelist = list(platinumTime["Token POS"].str.split(' ', expand=True).stack())
counts = pd.Series(platinumtimelist).value_counts()
platinumUniqueTime = counts.to_frame().reset_index() 
platinumUniqueTime.columns=['Token POS', 'Count']
platinumtimelist = list(platinumTime["Lemma POS"].str.split(' ', expand=True).stack())
counts = pd.Series(platinumtimelist).value_counts()
platinumUniqueLemmaTime = counts.to_frame().reset_index() 
platinumUniqueLemmaTime.columns=['Lemma POS', 'Count']

platinumTimeStats = pd.merge(platinumUniqueTime, platinumUniqueToken, on=['Token POS'], how='left')
colpos = platinumTimeStats["Token POS"].apply(lambda x:x.split("-")[-1])
platinumTimeStats = platinumTimeStats.join(colpos.to_frame(name="POS"))
platinumTimeLemmaStats = pd.merge(platinumUniqueLemmaTime, platinumUniqueLemma, on=['Lemma POS'], how='left')
colpos = platinumTimeLemmaStats["Lemma POS"].apply(lambda x:x.split("-")[-1])
platinumTimeLemmaStats = platinumTimeLemmaStats.join(colpos.to_frame(name="POS"))
platinumTimeStats.columns = ['Token POS', 'Count Time', 'Count Token', "POS"]
platinumTimeLemmaStats.columns = ['Lemma POS', 'Count Time Lemma', 'Count Token Lemma', "POS"]
platinumTimeStats = platinumTimeStats[(platinumTimeStats["POS"] == "NN") | (platinumTimeStats["POS"] == "RB") | (platinumTimeStats["POS"] == "JJ")]
platinumTimeLemmaStats = platinumTimeLemmaStats[(platinumTimeLemmaStats["POS"] == "NN") | (platinumTimeLemmaStats["POS"] == "RB") | (platinumTimeLemmaStats["POS"] == "JJ")]
```

```python
try:
    from itertools import izip as zip
except ImportError: # will be 3.x series
    pass
from itertools import product

def find_continuous_items(data):
    for p in product(*data):
        if all(b-a==1 for a, b in zip(p, p[1:])):
            yield p
```

```python
platinumTimeMultiTokens['Annotated'] = 0
platinumTimeMultiTokens['Not Annotated'] = 0

for i in range(len(platinumTimeMultiTokens)):
    elems = platinumTimeMultiTokens["Token"].iloc[i].split(" ")
    no_elem = len(elems)
    indexes = []
    
    for elem in elems:
        indexes.append(platinumTokens.index[platinumTokens['Lowercase Token'] == elem].tolist())
    
    occurrences = list(find_continuous_items(indexes))
    
    for occ in occurrences:
        occ = list(occ)
        subset = platinumTokens.iloc[occ[0]:occ[-1]+1]
        
        if sum(subset["Is Time Expression"]) == len(occ):
            platinumTimeMultiTokens['Annotated'].iloc[i] = platinumTimeMultiTokens['Annotated'].iloc[i] + 1
        else:
            platinumTimeMultiTokens['Not Annotated'].iloc[i] = platinumTimeMultiTokens['Not Annotated'].iloc[i] + 1
```
